## Data Preprocessing

```python
all_lyric = '\n'.join(df['lyrics'])
```

Tokenize the text to pass through the model

```python
import tensorflow as tf
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
#making an object
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts([all_lyric])
```

```python
total_tokens = len(tokenizer.word_index) + 1
input_sequence = []
```

```python
for line in all_lyric.split('\n'):
  token_list = tokenizer.texts_to_sequences([line])[0]
  for i in range(1, len(token_list)):
    n_gram_seq = token_list[:i+1]
    input_sequence.append(n_gram_seq)
```

Padding Sequences

```python
max_seq_len = max(len(x) for x in input_sequence)
```

```python
input_sequence = np.array(pad_sequences(input_sequence, maxlen = max_seq_len, padding = 'pre'))
```

Splitting into features and lables

```python
feature, label = input_sequence[:, :-1], input_sequence[:, -1]
```

```python
label = tf.keras.utils.to_categorical(label, num_classes = total_tokens)
```

Now, we have in handy our feature matrix and the labels required. We can proceed with constructing the model. We will form a sequential model with multiple layers.

## Building the Text Generation Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
model = Sequential([
    Embedding(total_tokens, 30, input_length = max_seq_len - 1),
    LSTM(30),
    Dense(total_tokens, activation = 'softmax')
])
```

```python
model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

```python
model.summary()
```

## Train The Model

```python
history = model.fit(feature, label, epochs = 50)
```
